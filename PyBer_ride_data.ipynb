```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
import statistics
```

```python
# Load in csv
data_to_load = os.path.join("Resources", "PyBer_ride_data.csv")
```

```python
pyber_ride_df = pd.read_csv(data_to_load)
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", color ="red")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()

print(x_axis)
print(tick_locations)
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
stdev = statistics.stdev(pyber_ride_df['Avg. Fare ($USD)'])
stdev
```

```python
pyber_ride_df.plot.bar(x='Month', y='Avg. Fare ($USD)', yerr=stdev, capsize=3, color="cyan")
plt.yticks(np.arange(0, 60, step = 5.0))
plt.ylabel('Avg. Fare ($USD)')
plt.show()
```
